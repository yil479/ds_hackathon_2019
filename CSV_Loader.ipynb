```python
import csv
import pandas as pd
import json
```

```python
data = "2019_Hackathon_Data.csv"
ad_data = "2019_Hackathon_Additional_Data.csv"
```

```python
file = pd.read_csv(data)
```

```python
with open(data) as f:
    reader = csv.reader(f, delimiter=",")
    rows = [row for row in reader]
    # original keys
    keys = rows[0][:-1]
    # rows of original values
    o_rows = [row[:-1] for row in rows[1:]]
    
    new_rows = []
    for k, row in enumerate(rows[1:]):
        row = row[-1] # get [...]
        row = row[1:-1] # remove [, ]
        items = row.split('},') # get out every {...}
        items = [item[1:] for item in items]
        items[-1] = items[-1][: -1]
        
        # calculate fraction sum
        t_sum = 0
        p_sum = 0
        
        for l, item in enumerate(items):
            infos = item.split(',')
            values = []
            
            for p, info in enumerate(infos):
                key, value = info.split(':')
                key = key[1:-1]
                if value[0] == '\"':
                    value = value[1:-1]
                value = float(value)
                if value > 1:
                    value = int(value)
                if p == 3:
                    value /= 1000
                values.append(value)
                if p == 1:
                    t_sum += value
                elif p == 2:
                    p_sum += value
                    
                if (k == 0) and (l == 0): # need to modify keys
                    keys.append(key)
            
            current_row = o_rows[k].copy()
            current_row += values
            new_rows.append(current_row)
        
#        print(t_sum)
#        print(p_sum)
    
    print(keys)
    print(new_rows[0])
            
            
        
    
```

```python
with open('mod_file.csv', mode='w') as file:
    f_writer = csv.writer(file, delimiter=',')

    f_writer.writerow(keys)
    for row in new_rows:
        f_writer.writerow(row)
```
